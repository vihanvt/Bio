```python
import py3Dmol
import ipywidgets as widgets
from IPython.display import display
import requests
#displaying using py3dmol, with better user interface

#dictionary containing ids of proteins extracted from PDB
pdbid = {
    "Hemoglobin": "1A3N",
    "Insulin": "4INS",
    "Keratin": "1A1X",
    "Histone": "3LEL",
    "Myoglobin": "1MBO",
    "Cytochrome C": "1HRC",
    "Collagen": "1BKV",
    "Elastase": "1ELA",
    "Lysozyme": "1LYZ",
    "Alcohol Dehydrogenase": "4W6Z",
    "RNA Polymerase": "4A3E",
    "Glutamine Synthetase": "2GLS",
    "Lactate Dehydrogenase": "6LDH",
    "ATP Synthase": "1E79",
    "Ribosome": "4V6C",
    "Ubiquitin": "1UBQ",
    "DNA Polymerase": "1BPX",
    "Actin": "1ATN",
    "Tubulin": "1TUB",
    "Ferritin": "1FHA",
    "Amylase": "1SMD",
    "Glucose Oxidase": "1GAL",
    "Chymotrypsin": "2CHA",
    "Trypsin": "2PTC",
    "Hexokinase": "1HKG"
}

viewer = py3Dmol.view(width=800, height=400)

def show_protein(change):
    pdb_id = pdbid[dropdown.value]
    viewer.clear()
    url = f"https://files.rcsb.org/view/{pdbid}.pdb"
    #here is the display code
    response = requests.get(url)
    if response.status_code == 200:
        pdb_data = response.text
        viewer.addModel(pdb_data, 'pdb')
        viewer.setStyle({'cartoon': {'color': 'spectrum'}})
        viewer.zoomTo()
        viewer.show()
    else:
        print(f"Error fetching PDB file for {pdbid}")

dropdown = widgets.Dropdown(
    options=pdbid.keys(),
    description='Select Protein:',
)
dropdown.observe(show_protein, names='value')
display(dropdown)
show_protein(None)#calling the function
```
